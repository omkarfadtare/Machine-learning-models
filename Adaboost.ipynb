## __Adaboost (Stagewise adaptive)__
- It is a supervised machine learning algorithm which can be used for both regression and classification problems.
- Adaboost is nothing but a Sequential + Gradual + Additive method.
- It is an ensemble algorithm that combines many weak learners (decision trees) and turns it into one strong learner.
- These weak learners in Adaboost are known as decision stumps. Adaboost is also reffered as forest of stumps.
- __Weak learner:__ A machine learning algirithm whose accuracy is just over 50%.
- __Decision stumps:__ Decision stumps area kind of weak learners or decision tree having max depth one and leaf nodes two.

## __Adabost Regression & Classification__
- Assign initial sample weights to each data point.


- Calculate gini value for each feature.


- Calculate gini index for each feature.


- Select a feature having lowest gini index and highest information gain to split the tree.
- Calculate total error (Total Error is equal to the sum of the weights of the incorrectly classified samples.)
- Calculate amount of say for the decision stump


- Calculate new sample weights for correctly and incorrectly classified data points for next decision stump
- We will decrease the new sample weight for correctly classified data points and increase the new sample weight for incorrctly classified data points in next decision stump.


- Calculate the new normalised sample weights

- Since, incorrectly classified data points have more weight than that of correctcly classified data points thus while bootstrapping missclassified data points are more likely to be selected multiple times, and thus, the next stump will focus more on classifying the misclassified data points correctly.
- In this way we will keep on building decision stumps by repeating same steps, until we have reached max number of decision stumps for adaboost by default value of decision stumps is 50.

__Hyperparameters of Adaboost__
- __`base_estimator`__ type of weak learner we want to use by default it uses decision tree
- __`n_estimators`__ the number of weak learners we want by default it is 50
- __`learning_rate`__ shrink the contribution of each classifier by default it is 1
