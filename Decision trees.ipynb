## __Decision Trees- Theory__
- Decision Tree is a supervised machine learning algorithm that can be used for both Regression and Classification problems.
- Decision trees are nothing but a giant structure of nested if-else conditions.
- Decision trees are nothing but hyperplane which run parellal to any one of the axis to cut your co-ordinate system into hyper cuboids.


__Types of Decision tree algorithms:__
- ID3 (Iterative Dichotomiser): which uses entropy to calculate information gain.
- C4.5 (Successor of ID3): which uses infomation gain ratio as attribute selection measure.
- CART (Classification and Regression Trees) which uses gini value/gini impurity to calculate gini index.

__Advantages of Decision trees__
- Intuitive and easy to understand
- Minimal data preprarion is required
- Requires less training time; The training time for decision trees is less as compared to ensemble techniques like Random Forest because it generates only one Tree unlike the forest of trees in the Random Forest.

__Disadvantages of Decision trees__
- Chances of Overfitting
- Prone to errors in imbalanced dataset
- Not suitable for large datasets; If the data size is large, then one single Tree may grow complex and lead to overfitting. So in this case, we should use Random Forest instead.

__Types of nodes in Decision trees__
- Root node: It is the top-most node of the Tree from where the Tree starts.
- Decision nodes
- Leaf nodes


- To select root node/decision node we have to calculate information gain which should be always maximum.
- To calculate information gain we can use either entropy or gini index which should be minimum.
- Less entropy or gini, maximum is the information gain. 

__Entropy:__ 
- It is nothing but the measure of disorder/impurity/purity of data.
- More the uncertainty in the data more is the entropy.
- For binary classification problems value of entropy lies in between (0-1)
- For multiclass classification problems minimum value of entropy is 0 and maximum will be more than 1.
- We can use both log2 or loge to calculate entropy.
- CART algorithm uses entropy to calculate information gain.


__Information gain:__ 
- It is a metric used to train the decision tree which basically measures the quality of split based on the decrease in entropy after a data set is split on an attribute. 
- Basically decision tree is all about finding attribute that returns the highest information gain.

__Gini index:__
- It also measures the disorder/impurity/purity of data.
- ID3 uses gini value to calcualte gini index.
- In sklearn, by defualt value of the criterian is gini.
- For binary classification the minimum value for gini is 0 and the maximum value will be 0.5.

### __Pruning__
- Pruning technique is used to prevent overfiting of the model.
- It is a process of cutting down the tree or removing sub-nodes of a decision nodes in such a way that it will reduce overfitting and can return acceptable accuray as well.

__Types of prunning__
- Pre-Pruning (Done after the construction of decision tree, so as to reduce the overfitting)
- Post-Pruning (Is done so a to prevent overfitting)

__ccp_alpha__
- ccp stands for cost complexity pruning
- Greater values of ccp_alpha increase the number of nodes pruned.

__Hyperparameters of Decision tree__
- __`max_depth:`__ the longest path between the root node and the leaf node.
- __`min_samples_leaf:`__ minimum number of samples that should be present in the leaf node after splitting a node. [Default = 1]
- __`min_sample_split:`__ minimum required number of observations in any given node to split it. [Default = 2]
- __`max_features:`__ number of maximum features provided to each tree in a random forest.
- __`criterian:`__ the function to measure the quality of a split. gini for the Gini impurity and entropy for the information gain. In case of regression it will be `mse` and `mae`
