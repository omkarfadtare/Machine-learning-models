## __Gardient boost__
- Gradient Boosting trains models in a Gradual, Additive and Sequential manner.
- The major difference between AdaBoost and Gradient Boosting Algorithm is how the two algorithms identify the shortcomings of weak learners.
- In adaboost, we increase the weightage of misclassified data points based on the output of previous of weak learner. In gradient boost we take smaller steps towards right decision by minimizing the residual.
- In adaboost we use decision stumps of max depth 1 and leaf nodes two, whereas in gradientboost we build a decision trees with max leaf nodes between (8-32)

## __GB Regression__
- Assign initial prediction for first decision leaf as average of target column.
- Depending upon the predictions of first model calculate pseudo residuals for each data point.
- __Pseudo residuals__ Pseudo residuals is based on the linear regression. The term pseudo is a reminder that we are dealing with gradient boost not linear regression.
- (Pseudo residual = Yactual - Ypredicted)
- Depending upon the information gain plot a complete decision tree for residuals of those data points.
- Calculate optput value. A leaf node having more than 1 residuals take avg.
- Calculate final predictions for model2
- (Final predictions = Initial prediction for decision leaf + (learning rate * Output value))
- Depending upon the new prediction find the pseudo residuals.
- In this way we will keep building models or decision trees until the residuals become comparatively small or have reached max number of decision trees.
- Basically, we take smaller steps towards right decision by minimizing the residual.

## __GB Classsification__
- Assign initial prediction for first decision leaf as model1 prediction, for that calculate log of odds of target column and convert it to probability using logistic function.


- Depending upon the predictions of first model calculate pseudo residuals for each data point.
- (Pseudo residual = Yactual - Ypredicted)
- Depending upon the information gain plot a complete decision tree for residuals of those data points.
- Calculate output value for each leaf node.

- Calculate final predictions for model2
- (Final predictions(which will be a log of odds value) = Log of odds for initial prediction for decision leaf + (learning rate * Output value))

- Convert log of odds of final prediction to probability using logistic function.

- Depending upon the new prediction find the pseudo residuals.
- In this way we will keep building models or decision trees until the residuals become comparatively small or have reached max number of decision trees.
- Basically, we take smaller steps towards right decision by minimizing the residual.

__Hyperparameters of Gradient boost__
- __`min_samples_split`__ minimum split required
- __`min_samples_leaf`__ minimum leaves required
- __`learning_rate`__ controls the magnitude of this change
- __`n_estimators`__  the number of weak learners we want by default it is 100
- __`subsample`__ the fraction of observations to be selected for each tree
