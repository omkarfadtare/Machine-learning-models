__What is Ensemble technique/learning in machine learning?__
- In simple words ensemble meas a group of something, but in machine learning point of view ensemble techniques is a group of machine learning models. Ensemble learning is based on the fact of wisdom of freedom. 
- For ensemble learning we have two ways either we use a group of different models on same dataset or we can use a group of same models with different dataset.

__Types of Ensemble learning__
- Voting classifier
- Stacking
- Bagging
- Boosting

__Pros and Cons of Ensemble technique__
- Since ensemble technique uses a group of machine learning models it will increase the computational time complexity of the model.
- It will increase the performance or accuracy of the model.
- It achieves low bias and low variance which reduces chances of overfitting of the model.
- Robust to any kind of data.

__When to use ensemble technique__
- Always. There is no reason to not to use ensemble technique.


### 1) Voting classifier
__Intuition of Voting classifier__
- Voting classifier is based upon the concept that a group of several different machine learning models are used on same dataset to get the final prediction.
- In voting classifier a group of different machine learning models are used as base models. And based upon the predictions made by each individual models, final prediction is made.
- In case of classification problems final prediction will be made by voting and in case of regression problems final prediction is made by aggregating all the predictions made by individual models or calculating the mean of all the predictions made by individual models.

__Assumptions of voting classifier__
- Model should be independent and minimum accuracy of the model should be more than 50%


__Types of voting classifier for classification__
- __Hard voting classifier:__ final prediction is made directly based up on voting
- __Soft voting classifier:__ final prediction is made based upon maximum probabilty of class. probabilities of each class is calculated and after aggregation of all probabilities, final prediction is made. 


### 2) Stacking
- Stacking is based upon the concept that a group of several different machine learning models are used on same dataset to get the final prediction.
- In stacking a group of different machine learning models are used as base models. And then the predictions made by each individual models are fed to another machine learning model(Genralizer model) as a input training data and the final prediction is made.


### 3) Bagging(Booststrap aggregating)
- Bagging is based upon the concept that a group of several same machine learning models are used on different dataset to get the final prediction.
- Bagging is an ensemble technique where we have a group of same base machine learning models provided with the different subset of data so as to train our models in different ways and then by aggregating(in case of regression) or voting classifier(in a case of classification) we can get our final prediction.
- Bagging is based upon the concept that the group of several same machine learning algorithms or models are used on different dataset to get the final prediction.
- When we are using decision tree as our base model for bagging, then it is known as Random forest.
- Bagging uses parellal approach. 
- Different dataset(bootstrapping of data)+ Group of same machine learning models + Voting classifier/Aggregating = Final prediction.
- __Booststraping__ is a method of sampling data for bagging ensemble learning. It is just a process of splitting our original dataset into number of sample datasets with/without duplicate entries having equal/not equal number of rows/columns. 
- __Bagging:__ sampling of rows with duplicates
- __Pasting:__ sampling of rows without duplicates
- __Random space:__ sampling of columns with or without duplicates
- __Random patches:__ sampling of both rows and columns with or without duplicates


__OOB score__
- OOB stands for out of bag samples, these are the samples or rows of the original dataset which do not get selected in bootstraping. We can use these samples as test datset to calculate our model accuracy.
- Generally 37% of samples from original dataset lies in OOB.


### 4) Boosting
- Boosting is based upon the concept that a group of several same machine learning models are used on different dataset to get the final prediction.
- Boosting uses __Sequential, Gradual, Stagewise additive method__.
- In boosting technique, we will keep building models or decision trees until the residuals become comparatively small or have reached max number of decision trees.
- Basically, we take smaller steps towards right decision by minimizing the residual.

__Statagewise Additive modeling:__
- Machine learning is nothing but finding relatioship between input and ouput variables and that relationship can be defined as mathematical functions.
- In some cases it is a stright line where you can apply linear regression sometimes, it is polynomial where you can apply polynomial regression but sometimes it is more complex which you can not explain using linear or polynomial. 
- Where you have to break down that complex function in smaller functions and then you can get your final function by adding those small functions this is called additive modeling. You do this stagewie so thats why it is stagewise additive method.

__Difference between Bagging & Boosting__
- In bagging low biased and high variance ml models(fully grown decision trees) are used while in boosting high biased and low variance ml models(shallow decision trees) are used.
- In bagging models are trained parellaly while in boosting models are trained sequentially.
- In bagging amount of say/weightage for each model is same while in boosting amount of say/weightage is different.
