## __Random forest__
- Random forest is a bagging ensemble technique and a supervised machine learning algorithm that can be used for both classification and regression problems. It is also known as forest of decision trees.
- Bagging is nothing but bootstrapping plus aggregation.
- Although we can use any machine laearning algorithm for bagging but decision tree performs better in this case, when we use decision tree as base model in bagging ensemble then it is known as Random forest.
- __Bootstrapping:__
- Bootstrapping means instead of training on all the observations, each decision tree of random forest is trained on a subset of the observations.
- Perform bootstrapping of data.

- Plot a complete decision tree based on each bootstrap.
- So as to get the final prediction for classification problems we use voting classifier and in a case of regression problem we basically calculate mean of all the outputs of the decison trees.
- When we have total 3 classifers in bagging out of which one classifier is wrong and two others are correct. By using voting classifier we can easily rely on other two to achieve a good accuracy.
- __Need of boosting technique__
- But when all classifiers are mistaken in the same region, bagging does not work properly. Thats why we need Boosting where each model should focus on where the previous classifier performed poorly.

__Hyperparameters of Random forest__
- __`max_depth:`__ the longest path between the root node and the leaf node.
- __`min_sample_split:`__ minimum required number of observations in any given node to split it. [Default = 2]
- __`max_leaf_nodes:`__ restricts the growth of the tree.
- __`min_samples_leaf:`__ minimum number of samples that should be present in the leaf node after splitting a node. [Default = 1]
- __`n_estimators:`__ number of trees in the forest.
- __`max_sample:`__ determines what fraction of the original dataset is given to any individual tree
- __`max_features:`__ number of maximum features provided to each tree in a random forest.
- __`bootstrap:`__ method for sampling data points (with or without replacement). [Default = True]
- __`criterion:`__ the function to measure the quality of a split. gini for the Gini impurity and entropy for the information gain.
