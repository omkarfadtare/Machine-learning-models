## Theory
__Significance and Intution behind Logisstic regression__
- Logistic regression is a probabilistic supervised machine learning algorithm to predict binary outcome.
- In binary logistic regression problem, basically we try to classify data points with the help of decision boundry line. To plot the optimal decision boundry, first we need to find the optimal coefficients/weights for that decision boundry line.
- For that, first we initialize and assign some random wights to the data points and iteratively keep on updating those weights till we achieve the lowest loss function. 
- Then we select those optimal weights for which the loss function is minimum to plot a decision boundry. In 3D case, decision boundry will become plane and in 4D it will become a hyperplane.

- To find the best optimal coeffiecients/weights of decision boundries, we need to iterartively keep on updating the initial random weights and analyze the loss function and select those weights for which loss function is minimum. 
- Unfortunatly we do not have any closed form solution to find out those coeffiecients/weights just like linear regression do, so we need to use gradient descent to do so.
- Gradient descent will initialize and assign some random weights to the data points and iteratively keep on updtaing the weights till it reaches a global minima point. It is a point at which loss function will be minimum.
- In logistic regression grdient descent will find new weights by partial differentiation of log loss function.

__Loss function:__
- The Likelyhood function is used as a loss function in logistic regression which should be maximum. We use multiplication of probabilities to calculate likelyhood but since the values of probabilities always lies between 0-1 the result will be so small for large number of rows. For ex likelyhood for model1 = (Pa * Pb * Pc) likelyhood for model2 = (Pa * Pb * Pc) model having maximum value is selected.
- So as to overcome this issue we used log and as we known that log of (ab) = log(a) + log (b). The equation becomes log(Pa) + log(Pb) + log(Pc). But log of any value between (0-1) is always a -ve value. So we took -ve logs. Initially we selects maximum likelyhood but since we have taken negative logs we have to select the minimum value.
- Summation of -ve logs of likelyhood functions is known as binary cross entropy or log loss function.


__Types of logistic regression:__
- Simple logistic regression (one dependent and one independent feature having 2 classes) W0 + W1X1 + W2X2 = 0
- Multiple logistic regression (one dependent and more than one independent features having 2 classes) W0 + W1X1 + W2X2 + W3X3 = 0
- Multiclass logistic regression (where we have more than 2 class in target feature)

__Why dont we use Logistic instead of Perceptron?__
- Decision boundry in logistic cuts data points more symmetrically than perceptron decision boundry. In simple words, Logistic regression helps us to acheive best decision boundry or machine learning model.

- Thats why we assign weights to the data points to pull decision boundry more towards clasiified and push away from missiclassified data points.

__Assumptions of logistic regression:__
1) `No multicollinearity:`  There should not be any linear relationship between two or more independent features.

2) `Linear relationship between continious independent and logit function:` There should be a linear relationship between continuous independent features and logit function.

3) `No outliers:` There should not be any outliers present in data.

4) `Large sample size:` Sample size of data should be sufficiently large.
